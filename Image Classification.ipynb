```python
import numpy as np
import pandas as pd
import os
import cv2
import numpy as np
from sklearn.utils import shuffle
import matplotlib.pyplot as plot
import matplotlib.gridspec as gridspec
from random import randint
from IPython.display import SVG
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(
                    rescale = 1/255,
                    shear_range = 0.2,
                    zoom_range = 0.2,
                    horizontal_flip = True,
                    vertical_flip = False)
training_set = train_datagen.flow_from_directory(r"C:/Users/Vivek/OneDrive/Desktop/Image Dataset/seg_train/seg_train",
                                                 target_size = (64, 64),
                                                 class_mode = 'categorical')
```

```python
test_datagen = ImageDataGenerator(rescale = 1/255)
test_set = test_datagen.flow_from_directory(r"C:/Users/Vivek/OneDrive/Desktop/Image Dataset/seg_test/seg_test",
                                            target_size = (64, 64),
                                            class_mode = 'categorical')
```

```python
training_set.class_indices
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Convolution2D, MaxPooling2D, Flatten, Dense
```

```python
cnn = Sequential()
cnn
```

```python
cnn.add(Convolution2D(filters=32, kernel_size=3, activation='relu', input_shape=(64,64,3)))
```

```python
cnn.add(MaxPooling2D(pool_size=(2,2)))
```

```python
cnn.add(Convolution2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(MaxPooling2D(pool_size=(2,2)))
```

```python
cnn.add(Flatten())
```

```python
cnn.add(Dense(units=256, activation='relu'))
cnn.add(Dense(units=128, activation='relu'))
```

```python
cnn.add(Dense(units=6, activation='softmax'))
```

```python
cnn.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
cnn.fit_generator(training_set, steps_per_epoch= len(training_set), validation_data = test_set, validation_steps=len(test_set), epochs = 10)
```

```python
pred_datagen = ImageDataGenerator(rescale = 1/255)
pred_set = pred_datagen.flow_from_directory(r"C:/Users/Vivek/OneDrive/Desktop/Image Dataset/seg_pred/seg_pred",
                                            target_size = (64, 64),
                                            class_mode = 'categorical')
```

```python
def get_images(directory):
    label = 0
    Images = []
    Labels = []
    for labels in os.listdir(directory):
        if labels == 'glacier':
            label = 2
        elif labels == 'sea':
            label = 4
        elif labels == 'buildings':
            label = 0
        elif labels == 'forest':
            label = 1
        elif labels == 'street':
            label = 5
        elif labels == 'mountain':
            label = 3
        for image_file in os.listdir(directory+r'/'+labels):
            image1 = image.load_img(directory +r'/'+ labels + r'/' + image_file, target_size=(64, 64))
            image1 = image.img_to_array(image1)
            image1 = np.expand_dims(image1, axis=0)
            Images.append(image1)
            Labels.append(label)
            
    return Images,Labels
```

```python
im, la = get_images(r"C:/Users/Vivek/OneDrive/Desktop/Image Dataset/seg_test/seg_test")
im
```

```python
def predict(im):
    pred_result=[]
    for i in im:
        pred = np.argmax(cnn.predict(i),axis=1)
        pred_result.append(pred[0])
    return pred_result
```

```python
pred = predict(im)
```
